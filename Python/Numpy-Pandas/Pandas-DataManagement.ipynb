```python
def df_aggr(in_dfn,analysis_vars,group_vars):
    
    "This function will agreegate the specific measures at given supplied hierarchy"
  
    print("Input Dataframe Name = ", in_dfn.name)
    print("Analysis Variables = ", analysis_vars)
    print("Grouping Variables = ", group_vars)    
    
    global data_aggr
    
    if str(analysis_vars).upper() == "ALL":        
        data_aggr = in_dfn.groupby(group_vars).sum()
    else:                    
        for key,value in in_dfn.iteritems():
            if key in analysis_vars:   
                var = [key]
                name = "mean_"+str(key)
                print(name)
                keep_vars = var + group_vars
                print(keep_vars)
                _temp_df = in_dfn[keep_vars]
                data_aggr[name] = _temp_df.groupby(group_vars).mean()          
                
    return data_aggr
```

```python
by_vars = ["charges"]
top_20_id = ins_df.sort_values(by = by_vars, ascending = False).reset_index(drop = True)
top_20_id = top_20_id[:20]
```

```python
top_20_id
```

```python
# Calculate the discount for top 20 id
# > 60,000 - 0.15
# Between 50,000 to 60,000 - 0.10
# Between 40,000 to 50,000 - 0.05
top_20_id["Discount"] = np.where(top_20_id[by_vars] >= 60000, 0.15,
                                  (np.where(top_20_id[by_vars] >= 50000, 0.10, 0.05)))
top_20_id
```

```python
top_20_id["charges_after_discount"] = top_20_id.charges * top_20_id.Discount
top_20_id
```

```python
#Merge
ins_df_new = ins_df.merge(top_20_id, how = "left").fillna(0) #merge and fill NA with Zero
ins_df_new
```

```python
ins_df_new["charges_after_discount"] = np.where(ins_df_new.Discount == 0, ins_df_new.charges, ins_df_new.charges_after_discount)
ins_df_new = ins_df_new.sort_values(by = ["charges"], ascending = False)
ins_df_new
```
